```python
# importing necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url ='https://companiesmarketcap.com/p'
dlist=[]
li=[]
```

```python
page=1
while page !=62 :
  url=f"https://companiesmarketcap.com/page/{page}/"
  response= requests.get(url)
  page_contents=response.text
  soup = BeautifulSoup(page_contents, 'html.parser')
  for link in soup.find_all('a',class_='tooltip-title') :
    if link.get('href')=='#' :
     pass
    else : dlist.append(url+link.get('href'))
  page=page+1  

```

```python
for link in dlist :
  r = requests.get(link, allow_redirects=True)
  open('file.csv', 'wb').write(r.content)
  df = pd.read_csv('/content/file.csv', index_col=None, header=0)
  li.append(df)
  

frame = pd.concat(li, axis=0, ignore_index=True)  
```
